```julia
using Plots
```

```julia
# 初期値の設定

t0 = 0.0
u0 = 1.0
Δt = 0.1
n = 100

#ルンゲクッタ法の定義
function RK(u)
   r1 = r(u)
   r2 = r(u + Δt/2 * r1)
   r3 = r(u + Δt/2 * r2)
   r4 = r(u + Δt * r3)
   return u + Δt * (r1 + 2*r2 + 2*r3 + r4)/6
end

r(u) = (sin(u))^(1.2)

u = u0 
u_sq = [ u ]

for n in 1:n
   u = RK(u)
   push!(u_sq, u)
end

t_sq = [ n*Δt for n in 0:n ]
plot(t_sq, u_sq, xaxis = "time t", yaxis = "u", leg = false)
```
